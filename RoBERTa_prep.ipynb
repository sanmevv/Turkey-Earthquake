```python
import tensorflow as tf
from transformers import RobertaTokenizerFast, TFRobertaForSequenceClassification
```

```python
model_name = 'roberta-large'
roberta_tokenizer = RobertaTokenizerFast.from_pretrained(model_name)
roberta_seq = TFRobertaForSequenceClassification.from_pretrained(model_name)
```

```python
roberta_seq.summary()
```

```python
# Contains unicode codes (e.g. emojis)?
for t in tweets:
    if 'U+' in str(t):
        print(t)
```

```python
import re
# Contains HTML character entities?
for t in tweets:
    if '&' in re.sub(r'(&amp|&gt|&lt)','', str(t)):
        print(t)
```

```python
tweets = [t for t in tweets if isinstance(t, str)]

for t in tweets:
    if any([x in t for x in [' btw ',' omg ',' lol ',' thx ']]):
        print(t)
```

```python
def process_tweets(tweets):
    r = tweets
    r = [re.sub(r'https?://t.co/\w+','',t) for t in r]
    r = [re.sub('&amp;','&',t) for t in r]
    r = [re.sub('&gt;','gt',t) for t in r]
    r = [re.sub('&lt;','lt',t) for t in r]
    return r

tweets = process_tweets(tweets)
```

```python
temp = roberta_tokenizer(tweets[:5],padding='max_length',max_length=50)
temp.keys()
```

```python
# Combine all data in a separate list used only for determining the length of sequences
all_tweets = pd.concat([df,df_test],axis=0)['Tweet']
processed_tweets = process_tweets(all_tweets.astype(str))

# Comment out this line if you don't apply any pre-processing
all_tweets = process_tweets(all_tweets.astype(str))
max_len = max([len(t) for t in roberta_tokenizer(all_tweets)])
print(max_len)
```

```python
from transformers import RobertaTokenizer
```

```python
import numpy
from sklearn.model_selection import train_test_split
y = y[:-3]

X_train, X_test, y_train, y_test = train_test_split(tweets, y, test_size=0.30, random_state=42)
max_len = 128 # set maximum length of tweets
roberta_tokenizer = RobertaTokenizer.from_pretrained('roberta-base')

# tokenize tweets and pad/truncate them to max_len
X_train = roberta_tokenizer(X_train, padding='max_length', truncation=True, max_length=max_len, return_tensors='tf')
X_test = roberta_tokenizer(X_test, padding='max_length', truncation=True, max_length=max_len, return_tensors='tf')

# convert y labels to TensorFlow tensors
y_train = tf.convert_to_tensor(y_train)
y_test = tf.convert_to_tensor(y_test)
```

```python
print(y.sum()/len(y))
```

```python
y_train_sum = tf.reduce_sum(y_train).numpy()
y_train_len = len(y_train)

print(y_train_sum / y_train_len)
```

```python
y_test_sum = tf.reduce_sum(y_test).numpy()
y_test_len = len(y_test)
print(y_test_sum/y_test_len)
```

```python
batch_size = 8
train_dataset = tf.data.Dataset.from_tensor_slices((dict(X_train),y_train))
train_dataset = train_dataset.batch(batch_size)
test_dataset = tf.data.Dataset.from_tensor_slices((dict(X_test),y_test))
test_dataset = test_dataset.batch(batch_size)
```

```python
print(train_dataset)
print(test_dataset)
```

```python
temp_x, temp_y = next(iter(test_dataset))
temp = roberta_seq(temp_x,temp_y)
temp
```

Fine-tuning RoBERTa

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=5e-6)
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
roberta_seq.compile(optimizer=optimizer,loss=loss,metrics=['accuracy'])
```
